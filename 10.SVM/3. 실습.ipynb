#1. 데이터 셋 읽기
> iris 데이터를 읽어 들이고 data와 label을 dataframe으로 만드시오

```python
from sklearn.datasets import load_iris

iris = load_iris()
```

```python
import pandas as pd

df_data = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df_labels = pd.DataFrame(iris.target)
df_data.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.scatter(np.array(df_data)[:,0], np.array(df_data)[:,1], c=df_labels)
plt.show()
```

#2. 비선형 분류

```python
import sklearn.svm as svm

svm_clf = svm.SVC(kernel='rbf')
```

> 학습데이터와 검증데이터를 나누고 모델을 만든 후 정확도를 산출하시오

```python
from sklearn.model_selection import train_test_split

y_labels = df_labels.values.reshape(-1)
X_train, X_test, y_train, y_test = train_test_split(df_data, y_labels, test_size=0.2, random_state=62)
```

```python
svm_clf.fit(X_train, y_train)
svm_clf.score(X_test, y_test)
```

#3. scaling

##3.1 표준화
> 표준화를 이용하여 정확도를 산출하시오

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(df_data)
standard_scaled = scaler.transform(df_data)
standard_scaled[:2]
```

```python
import sklearn.svm as svm

svm_clf = svm.SVC(kernel='rbf')
```

```python
from sklearn.model_selection import train_test_split

y_labels = df_labels.values.reshape(-1)
X_train, X_test, y_train, y_test = train_test_split(standard_scaled, y_labels, test_size=0.2, random_state=62)
```

```python
svm_clf.fit(X_train, y_train)
svm_clf.score(X_test, y_test)
```

##3.2 정규화
> 정규화를 이용하여 정확도를 산출하시오

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(df_data)
minmax_scaled = scaler.transform(df_data)
minmax_scaled[:2]
```

```python
import sklearn.svm as svm

svm_clf = svm.SVC(kernel='rbf')
```

```python
from sklearn.model_selection import train_test_split

y_labels = df_labels.values.reshape(-1)
X_train, X_test, y_train, y_test = train_test_split(minmax_scaled, y_labels, test_size=0.2, random_state=62)
```

```python
svm_clf.fit(X_train, y_train)
svm_clf.score(X_test, y_test)
```

#4. 교차 검증

```python
import sklearn.svm as svm

svm_clf = svm.SVC(kernel='rbf')
```
